```python
import bs4
from tqdm import tqdm
import pandas as pd
import datetime
```

```python
f = open("rbc.txt", "r", encoding="utf-8")
code = f.read()
f.close()
```

```python
df = pd.DataFrame(columns=["url", "date", "title", "url_pic", "body", "tickers"])
df
```

```python
soup = bs4.BeautifulSoup(code, "html.parser")
```

```python
def date(s):
    if "," in s:
        return s
    else:
        return "10 июл, " + s
```

```python
def change_tickers(soup: bs4.BeautifulSoup) -> str:
    d = {}
    for i in soup.find_all("div"):
        d[i.a["data-companyshortname"]] = (i.a["data-exchangepercent"] if i.a["data-exchangepercent"] else "0") + "%"
    return str(d)
```

```python
for i in tqdm(soup.find_all("div", attrs={"class": "q-item js-rm-central-column-item js-load-item overflow-visible"})[1:]):
    try:
        next = i.find_all("a")
        l = [
            next[0]["href"], 
            date(i.find("span", attrs={"class": "q-item__date__text"}).text.replace("\n", "").strip(" ")),
            next[0].span.text.replace("\n", "").strip(" "),
            i.find("img")["src"],
            i.find("span", attrs={"class": "q-item__description"}).text.replace("\n", "").strip(" "),
            change_tickers(i.find("div", attrs={"class": "swiper-wrapper"}))
            ]
        df.loc[len(df.index )] = l
    except BaseException:
        next = i.find_all("a")
        l = [
            next[0]["href"], 
            date(i.find("span", attrs={"class": "q-item__date__text"}).text.replace("\n", "").strip(" ")),
            next[0].span.text.replace("\n", "").strip(" "),
            None,
            i.find("span", attrs={"class": "q-item__description"}).text.replace("\n", "").strip(" "),
            change_tickers(i.find("div", attrs={"class": "swiper-wrapper"}))
            ]
        df.loc[len(df.index )] = l

    
```
